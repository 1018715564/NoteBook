```python
class wine_data_analysis(object):
    def __init__(self, input_data):
        pass
        
        
    # 数据预处理函数，建立合格的数据模型
    def data_preprocess(self):
        global data1
        # 数据去重
        data1.drop_duplicates(inplace=True)
        # 价格数据列的数据逻辑性检验
        data1.drop(index=data1[data1['价格']<=0].index.tolist(), inplace=True)
        # 用特殊值‘-1’对‘评论数’数据列进行补值
        data1.loc[data1['评论数'].isnull(),'评论数'] = -1
        answer1=tkinter.messagebox.askokcancel('提示', '数据预处理完毕，继续点确定')
        
        
    # 产品季节性分析   
    def prod_seasonal_analysis(self):
        global data1
        # 画布分割
        plt.figure(0,dpi=150, figsize=(30, 21))
        gs = gridspec.GridSpec(4, 4)
        pic_1 = plt.subplot(gs[0, 0])
        pic_2 = plt.subplot(gs[1, 0])
        pic_3 = plt.subplot(gs[2, 0])
        pic_4 = plt.subplot(gs[3, 0])
        pic_5 = plt.subplot(gs[3, 1])
        pic_6 = plt.subplot(gs[3, 2])
        pic_7 = plt.subplot(gs[3, 3])
        pic_8 = plt.subplot(gs[2, 3])
        pic_9 = plt.subplot(gs[1, 3])
        pic_10 = plt.subplot(gs[0, 3])
        pic_total = plt.subplot(gs[0:3, 1:3])
        for pics, brand_list in zip([pic_1, pic_2, pic_3, pic_4, pic_5, pic_6, pic_7, pic_8, pic_9, pic_10], 
                                    np.array(list(set(data1['品牌']))).reshape(10, 5)):
            for brands in brand_list:
                (data1.loc[data1['品牌'] == brands]).groupby('日期')['近30天销量（瓶）'].sum().plot(ax=pics, label=brands, legend=True, alpha=0.5)
        data1_total = data1.groupby('日期')['近30天销量（瓶）'].sum()
        data1_total.plot(kind='line', ax=pic_total, label='各类白酒总销量', legend=True, alpha=0.5)
        plt.bar([str(date) for date in data1_total.index], data1_total.tolist(), width=8)
        plt.title('各品类产品销售趋势图和所有品类总销售趋势图')
        tkinter.messagebox.showinfo('结论：', '(1)产品销量几乎都与季节十分相关，春、冬两季是白酒销售旺季，夏、秋相反。\n(2)产品不同，销量波动幅度有所不同，但波动几乎同步。')
    
    
    # 产品潜力分析
    def potential_prod(self):
        global data1
        # 首先对产品的搜索次数进行排名
        data_groupby1 = data1.groupby('品牌')['近30天销量（瓶）', '评论数', '被搜索次数'].sum()
        # 选择搜索次数排名前15名的产品
        data_groupby2 = data_groupby1.sort_values('被搜索次数', ascending=False).iloc[:20, :]
        potential_prod_name = (data_groupby2.sort_values(['评论数', '近30天销量（瓶）']).iloc[:12,:].index).tolist()
        tkinter.messagebox.showinfo('结论：', '潜力产品是：' + ','.join(potential_prod_name))
        
        
    # 带来销量的产品    
    def main_sold_product(self):
        global data1
        # 确定带来大部分销量的产品
        main_sold_prod = data1.groupby('品牌')['近30天销量（瓶）'].sum().sort_values(ascending=False)[:10].index.tolist()
        # 主销产品品类的总销量占所有产品品类总销量的比例
        main_prod_proportion = str(np.round(data1.groupby('品牌')['近30天销量（瓶）'].sum().sort_values(ascending=False)[:10].sum()/  
                               (data1.groupby('品牌')['近30天销量（瓶）'].sum()).sum(),4)*100) + '%'
        tkinter.messagebox.showinfo('结论：', '带来销量的主要产品是：' + ','.join(main_sold_prod) 
                                    + ',这些主销产品带来的销量占总销量的' + main_prod_proportion)
        
        
    # 确定销售变化起伏大的产品    
    def sales_hight_var_prod(self):
        global data1
        sales_highvar_prod = data1.groupby('品牌')['近30天销量（瓶）'].std().sort_values(ascending=False)[:9].index.tolist()
        pic1, axes1 = plt.subplots(3, 3, dpi=120, figsize=(16, 12))
        for pic_index, item_prods in zip([(0, 0), (0, 1), (0, 2), (1, 0), (1, 1), (1, 2), (2, 0), (2, 1), (2, 2)], sales_highvar_prod):
            data1.loc[data1['品牌'] == item_prods].groupby('日期')['近30天销量（瓶）'].sum() \
            .plot(legend=True, label=item_prods, ax=axes1[pic_index[0], pic_index[1]])
        tkinter.messagebox.showinfo('结论：', '销售变化起伏大的产品是：' + ','.join(sales_highvar_prod))
        
    
    # 最受欢迎的价格范围
    def price_volume(self):
        global data1
        # 组名
        group_name = ['(' + str(intval.left) + ', ' + str(intval.right) + ']' for intval in pd.cut(data1['价格'], bins=10, 
                      right=False, include_lowest=True).values.categories.tolist()]
        data2 = data1.copy()
        data2['价格'] = pd.cut(data1['价格'], bins=10, right=False, include_lowest=True).values.codes
        # 组高
        hight1 = data2.groupby('价格')['近30天销量（瓶）'].sum().reindex(range(10), fill_value=0).tolist()
        Sales = pd.Series({name : value for name, value in zip(group_name, hight1)})
        pic2, axes2 = plt.subplots(1, 1, dpi=120, figsize=(20, 8))
        axes2.bar(group_name, hight1)
        tkinter.messagebox.showinfo('结论：', str(group_name[0]) + ' : ' + str(Sales[0]) + '\n' + str(group_name[1]) + ' : ' + str(Sales[1]) + '\n' + str(group_name[2]) + ' : ' + str(Sales[2]) + '\n' + str(group_name[3]) + ' : ' + str(Sales[3])
                                    + '\n' + str(group_name[4]) + ' : ' + str(Sales[4]) + '\n' + str(group_name[5]) + ' : ' + str(Sales[5]) + '\n' + str(group_name[6]) + ' : ' + str(Sales[6]) + '\n' + str(group_name[7]) + ' : ' + str(Sales[7])
                                    + '\n' + str(group_name[8]) + ' : ' + str(Sales[8]) + '\n' + str(group_name[9]) + ' : ' + str(Sales[9]))
     
    
    # 最受欢迎的产品规格   
    def best_product_specification(self):
        global data1
        tkinter.messagebox.showinfo('提示：', '即将输出各产品最受欢迎的规格，请点击确定输出')
        print( data1.groupby(['品牌'])['近30天销量（瓶）', '商品名称'].max().loc[:, '商品名称' : '商品名称'])
    
```
